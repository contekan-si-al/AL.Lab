#### [Mencari customer yang mengganti layanan](https://academy.dqlab.id/main/projectcode/292/558/2793?pr=0)

```python
%%sql

SELECT
    t1.name,
    GROUP_CONCAT(t3.product_name)
FROM
    customer t1
    JOIN subscription t2 ON t1.id = t2.customer_id
    JOIN product t3 ON t2.product_id = t3.id
WHERE
    t1.id IN (
        SELECT
            customer_id
        FROM
            subscription
        GROUP BY
            customer_id
        HAVING
            count(1) > 1
    )
GROUP BY
    t1.name;
```
