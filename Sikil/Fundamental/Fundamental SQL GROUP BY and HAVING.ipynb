## Penggunaan Having

#### [Contoh penggunaan HAVING](https://academy.dqlab.id/main/livecode/291/555/2782?pr=0)

```python
%%sql

SELECT
  customer_id
FROM
  subscription
GROUP BY
  customer_id
HAVING
  COUNT(customer_id) > 1;
```

#### [Menampilkan Konsumen yang berubah berlangganan](https://academy.dqlab.id/main/livecode/291/555/2783?pr=0)

```python
%%sql

SELECT 
  customer_id,
  product_id,
  subscription_date
FROM
  subscription 
WHERE
  customer_id IN 
    (
     SELECT 
    	customer_id 
     FROM
      subscription 
     GROUP BY
      customer_id 
     HAVING
      COUNT(customer_id) > 1
  	) 
ORDER BY
  customer_id ASC;
```

#### [Menampilkan detail konsumen](https://academy.dqlab.id/main/livecode/291/555/2784?pr=0)

```python
%%sql

SELECT
  b.name,
	b.address,
  b.phone, 
  a.product_id, 
  a.subscription_date 
FROM
  subscription a 
JOIN
  customer b 
ON
  a.customer_id = b.id
WHERE
  b.id IN
  (
	SELECT
    customer_id 
  FROM
    subscription 
  GROUP BY
    customer_id 
  HAVING
    COUNT(customer_id) > 1
    ) 
ORDER BY
  b.id ASC;
```

## Penggunaan MAX, MIN dan AVG dalam Having

#### [Penggunaan Fungsi MAX pada Having](https://academy.dqlab.id/main/livecode/291/556/2786?pr=0)

```python
%%sql

SELECT
  product_id,
  MAX(total_price) AS total
FROM
  invoice
GROUP BY
  product_id;
```

```python
%%sql

SELECT
  product_id,
  MAX(total_price) AS total
FROM
  invoice
GROUP BY
  product_id
HAVING
  MAX(total_price) > 1000000;
```

```python
%%sql

SELECT
  product_id,
  MAX(pinalty) AS total
FROM
  invoice
GROUP BY
  product_id
HAVING
  MAX(pinalty) > 30000;
```

#### [Penggunaan Fungsi MIN pada Having](https://academy.dqlab.id/main/livecode/291/556/2787?pr=0)
